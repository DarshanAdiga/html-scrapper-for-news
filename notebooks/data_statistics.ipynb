```python
month_map_dict = {
    'ಜನವರಿ': 'January',
    'ಫೆಬ್ರವರಿ': 'February',
    'ಮಾರ್ಚ್': 'March',
    'ಏಪ್ರಿಲ್': 'April',
    'ಮೇ': 'May',
    'ಜೂನ್': 'June',
    'ಜುಲೈ': 'July',
    'ಜೂಲೈ': 'July',
    'ಆಗಸ್ಟ್': 'August',
    'ಸೆಪ್ಟೆಂಬರ್': ' September',
    'ಅಕ್ಟೋಬರ್': 'October',
    'ನವೆಂಬರ್': 'November',
    'ಡಿಸೆಂಬರ್': 'December'
}
```

```python
import dateutil.parser
def date_parser(date_str):
    if date_str is None:
        return None
    date_str = str(date_str).strip()
    for key,val in month_map_dict.items():
        if key in date_str:
            date_str = date_str.replace(key, val)
            break
    # Return the parsed date
    return dateutil.parser.parse(date_str)

df_articles['publish_date_dt'] = df_articles['publish_date'].map(lambda dt: date_parser(dt))
df_articles['publish_date_dt']
```

```python
# Convert to string
df_articles['publish_date_dt_str'] = df_articles['publish_date_dt'].map(lambda dt: dt.strftime("%Y-%m-%d %H:%M:%S %Z") if dt is not None else None)
```

```python
# Preprocess Keywords
#df_articles['keywords'].head(1000).values

def clean_keywords(kw_str):
    if kw_str is None:
        return None
    keywords = set()
    for kw in kw_str.split(','):
        kw = kw.strip()
        if len(kw) != 0:
            keywords.add(kw)
    if len(keywords) > 0:
        return ','.join(keywords)
    else:
        return None

df_articles['keywords_clean'] = df_articles['keywords'].map(lambda kw_str: clean_keywords(kw_str))
df_articles['keywords_clean'].head()
```

```python
#:: Unique count of Keywords
keyword_set = set()
for kw in df_articles['keywords_clean'].values:
    if kw is not None:
        keyword_set.update(kw.split(','))

# len(keyword_set)
# Exclude single character words
len([kw for kw in keyword_set if len(kw) > 1])
```

```python
# Clean Title & Description
def clean_title_description(text):
    if text is None:
        return None
    text = text.strip()
    if len(text) <= 1:
        return None
    else:
        return text

df_articles['title_clean'] = df_articles['title'].map(lambda tt: clean_title_description(tt))
df_articles['description_clean'] = df_articles['description'].map(lambda desc: clean_title_description(desc))
```

```python
from collections import Counter
def get_frequent_ngrams(df_data, column, n_gram=1):
    ngram_counter = Counter()
    total_count = 0
    for text in df_data[column].dropna().values:
        tokens = text.split()
        ngrams = zip(*[tokens[i:] for i in range(n_gram)])
        for ngram_word in [' '.join(ngram) for ngram in ngrams]:
            ngram_counter[ngram_word] += 1
            total_count += 1
    return ngram_counter,total_count
```

```python
# Sentence splitter
# iNLTK sentence_split
# https://indic-nlp-library.readthedocs.io/en/latest/indicnlp.tokenize.html
from indicnlp.tokenize import sentence_tokenize
import os

def indicnlp_splitter(text):
    return sentence_tokenize.sentence_split(text, lang='kn')

# indicnlp_splitter(text)
def save_sentences(df_data, base_dir_path):
    if not os.path.exists(base_dir_path):
        os.makedirs(base_dir_path)
        
    index_file = base_dir_path + 'index.csv'
    sentence_file = base_dir_path + 'sentences.csv'
    
    df_url_index = df_data[['url']]
    df_url_index = df_url_index.reset_index()
    # Save the index file
    df_url_index.to_csv(index_file, index=False)
    print('Saved the index file at {}'.format(index_file))
    
    df_art_text = df_data[['article_text']]
    all_sents = []
    # Iterate through the texts and segment the sentences
    for i,row in df_art_text.iterrows():
        text = row['article_text']
        sents = indicnlp_splitter(text)
        sents = [{'index':i, 'sentence':s} for s in sents]
        all_sents = all_sents + sents
    
    # Create a dataframe and save it
    df_sentence = pd.DataFrame(all_sents)
    df_sentence.to_csv(sentence_file, index=False)
    print('Saved the sentence file at {}'.format(sentence_file))
```

```python
# Generate and save the sentences
sent_dir_path = base_dir + 'processed/run2/'
save_sentences(df_articles, sent_dir_path)
```
